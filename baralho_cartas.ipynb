# Manipulando um baralho de cartas

>Este código demonstra como utilizar funcionalidades básicas como slicing, iteração, e embaralhamento em um baralho de cartas representado por uma classe Python.

```python
import collections
import random
```

```python
''' Um namedtuple chamado PlayingCard é criado para representar uma
carta de baralho com dois atributos: suit (naipe) e rank (valor). '''

# Definição de um namedtuple indicando como se constitui um baralho de cartas
Carta_Jogo = collections.namedtuple('cartas', ['naipe', 'valor'])

# Definimos a classe de objetos do baralho de cartas
class Baralho_Jogo:

  # Definimos os atributos do baralho
  valor = [str(valor) for valor in range(2, 11)] + ['J', 'Q', 'K', 'A']
  naipe = ['Spades', 'Diamonds', 'Hearts', 'Clubs']

  # Método Construtor do Baralho do Algoritmo
  def __init__(self):
    self.cartas = [Carta_Jogo(valor, naipe)
    for naipe in self.naipe
      for valor in self.valor]

  # Método para Obter o Comprimento do Baralho
  def __len__(self):
    return len(self.cartas)

  # Método para Obter uma Carta Específica ('__getitem__' = pegar item)
  def __getitem__(self, position):
    return self.cartas[position]

if __name__ == '__main__':
  baralho = Baralho_Jogo()

  # 1. SLICING DECK
  primeiro_corte = baralho[:4]

  # 2. ITERATING THROUGH A DECK
  print('A iteração inicial é:')
  for cartas in baralho:
    print(cartas)

  #3. INTERATING THROUGH THE DECK IN REVERSE
  print('\nA iteração reversa é:')
  for cartas in reversed(baralho):
    print(cartas)

  #4. SHUFFLING THE DECK
  print('\nO baralho embaralhado é:')
  lista_baralho = list(baralho)
  random.shuffle(lista_baralho)  # Shuffle the list
  for cartas in lista_baralho:
    print(cartas)
```
